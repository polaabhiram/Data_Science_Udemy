## Stemming 

Stemming is a process in which word is converted to its root stem ususally by eliminating suffixes or patterns

Variations of Porter stemmer are used for suffix stripping where as RegExp stemmer is used for trimming patterns(not definitely suffix but any pattern)

### Porter Stemmer

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
words=["eat","eaten","eats","history","sportfully","happiness","hardly","responses"]
for word in words:
    print(f"{word}:{stemmer.stem(word=word)}")
```

### Snowball Stemmer -Porter2Stemmer

```python
from nltk.stem import SnowballStemmer
stemmer=SnowballStemmer(language="english")

for word in words:
    print(f"{word}:{stemmer.stem(word=word)}")
```

### RegExp Stemmer

```python
from nltk.stem import RegexpStemmer
stemmer=RegexpStemmer('ing$|s$|e$|able$',min=4)

for word in words:
    print(f"{word}:{stemmer.stem(word=word)}")
```

## Lemmatization

### WordNet lemmatizer

Wordnet lemmatizer uses wordnet corpus that consists of semantic relations like synonys, antonyms, hyponyms

```python
from nltk.stem import WordNetLemmatizer
wordnet=WordNetLemmatizer()

for word in words:
    print(f"{word}:{wordnet.lemmatize(word=word,pos='n')}")
```

```python
from nltk.stem import WordNetLemmatizer
wordnet=WordNetLemmatizer()

for word in words:
    print(f"{word}:{wordnet.lemmatize(word=word,pos='v')}")
```

## Stop Words Elimination

Stopwords are frequently occurring words (such as “the”, “is”, “in”, and “and”) that often contribute less to distinguishing the meaning of a sentence in certain computational tasks. While removing them may make a sentence seem incomplete or less natural to humans, in some traditional text-processing methods they are treated as noise to reduce complexity and improve efficiency. However, they are not always unimportant, as in many cases they carry essential grammatical or semantic meaning and should be retained depending on the task.

In some cases, stopwords can help determine the meaning of a sentence, so it is recommended to use a custom set of stopwords based on the data and the use case. However, NLTK provides a standard list of stopwords.

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords.words("english")
```

```python
#Demonstrarion of eliminating stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer


stemmer=SnowballStemmer(language="english")
lemm=WordNetLemmatizer()
para="Natural language processing enables computers to understand and interpret human language. It is widely used in applications such as chatbots, sentiment analysis, and machine translation. By analyzing textual data, organizations can gain valuable insights, automate tasks, and improve decision-making processes."
words=word_tokenize(para)

no_stop=[word for word in words if word not in stopwords.words("english")]
stemmed=[stemmer.stem(word) for word in no_stop]
lemmatized=[lemm.lemmatize(word) for word in no_stop]
print(words)
print(no_stop)
print(stemmed)
print(lemmatized)
```
