```python
from sklearn.preprocessing import OneHotEncoder
oh=OneHotEncoder()

b=oh.fit_transform(df[['Airline']]).toarray()
Airlines=pd.DataFrame(b,columns=[oh.get_feature_names_out(['Airline'])])
Airlines
```

```python
df_copy=df.copy()
```

```python
df_copy.head()
```

```python
df_copy=pd.concat([df_copy,Airlines],axis=1)
```

```python
df_copy.drop(columns=['Airline'],inplace=True)
```

```python
df_copy.info()
```

```python
df_copy['Date_of_Journey']
df_copy['Date']=df_copy['Date_of_Journey'].str.split('/').str[0]
df_copy['Month']=df_copy['Date_of_Journey'].str.split('/').str[1]
df_copy['Year']=df_copy['Date_of_Journey'].str.split('/').str[2]
df_copy[['Date_of_Journey','Date','Month','Year']].head()

df_copy.drop(columns=['Date_of_Journey'],inplace=True)
```

```python
df_copy['Date']=df_copy['Date'].astype(int)
df_copy['Month']=df_copy['Month'].astype(int)
df_copy['Year']=df_copy['Year'].astype(int)
df_copy.info() 
```

```python
df_copy.head()
```

```python

df_copy['Arrival_Hour'] = df_copy['Arrival_Time'].str.split(':').str[0]
df_copy['Arrival_Min'] = df_copy['Arrival_Time'].str.split(':').str[1].str.split(' ').str[0]




```

```python
df_copy.head()
df_copy.info()
```

```python
df_copy['Arrival_Hour']=df_copy['Arrival_Hour'].astype(int)
df_copy['Arrival_Min']=df_copy['Arrival_Min'].astype(int)
df_copy.info()
```

```python



def parse_duration(x):
    if pd.isna(x):
        return 0, 0
    s = str(x).strip()
    hrs = 0
    mins = 0
    if 'h' in s:
        parts = s.split('h')
        try:
            hrs = int(parts[0].strip())
        except:
            hrs = 0
        if len(parts) > 1 and 'm' in parts[1]:
            try:
                mins = int(parts[1].replace('m','').strip())
            except:
                mins = 0
    elif 'm' in s:
        try:
            mins = int(s.replace('m','').strip())
        except:
            mins = 0
    return hrs, mins

dur_hours = []
dur_mins = []
for d in df_copy['Duration']:
    h, m = parse_duration(d)
    dur_hours.append(h)
    dur_mins.append(m)
df_copy['Duration_hours'] = dur_hours
df_copy['Duration_mins'] = dur_mins

df_copy['Total_Stops'] = df_copy['Total_Stops'].astype(str).str.replace('non-stop', '0 stops').str.replace('non stop', '0 stops').str.replace('nonstop', '0 stops')
df_copy['Total_Stops_Num'] = df_copy['Total_Stops'].str.extract('(\d+)').astype(float)

df_copy['Additional_Info'] = df_copy['Additional_Info'].fillna('No info')
df_copy['Route'] = df_copy['Route'].fillna('')
df_copy['Route_count'] = df_copy['Route'].apply(lambda x: len(x.split(' → ')) if x != '' else 0)

airline_dummies = pd.get_dummies(df_copy['Airline'], prefix='Airline')
source_dummies = pd.get_dummies(df_copy['Source'], prefix='Source')
destination_dummies = pd.get_dummies(df_copy['Destination'], prefix='Destination')

df_copy = pd.concat([df_copy, airline_dummies, source_dummies, destination_dummies], axis=1)

drop_cols = ['Airline', 'Source', 'Destination', 'Route', 'Dep_Time', 'Arrival_Time', 'Duration', 'Total_Stops']
for col in drop_cols:
    if col in df_copy.columns:
        df_copy.drop(columns=[col], inplace=True)

df_copy['Price'] = pd.to_numeric(df_copy['Price'], errors='coerce')
df_copy = df_copy.reset_index(drop=True)

df_copy.to_csv('./flight_price_cleaned.csv', index=False)
```
