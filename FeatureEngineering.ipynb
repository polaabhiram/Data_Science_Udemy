```python
import pandas as pd
import numpy as np
```

```python
n_0=np.random.normal(loc=0,scale=1,size=900)
n_1=np.random.normal(loc=0,scale=1,size=900)

class_0=pd.DataFrame({'f_1':n_0,'f_2':n_1})


class_0['target']=0
print(class_0.head())

```

```python
import numpy as np
import pandas as pd

n_0 = np.random.normal(scale=1, loc=1, size=100)
n_1 = np.random.normal(scale=1, loc=1, size=100)

class_1 = pd.DataFrame({'f_1': n_0, 'f_2': n_1})
class_1['target']=1
print(class_1)

```

```python
df=pd.concat([class_0,class_1])
df.tail()
```

```python

from sklearn.utils import resample

majority = df[df['target'] == 0]
minority = df[df['target'] == 1]

minority

```

```python
!pip install scikit-learn
```

```python
#Upsampling
upsampled=resample(minority,n_samples=len(majority),random_state=42,replace=True)
upsampled
```

```python
#Downsampling

down_sampled=resample(majority,n_samples=len(minority),replace=False,random_state=42)
down_sampled.reset_index().drop(columns=['index'],axis=1)
```

```python
#SMOTE
from sklearn.datasets import make_classification

X,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,
                   weights=[0.90],random_state=12)
```

```python
import pandas as pd
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
!pip install seaborn
```

```python
from imblearn.over_sampling import SMOTE


oversample=SMOTE()

X,y=oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```

```python
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
oversample_df=pd.concat([df1,df2],axis=1)

oversample_df['target'].value_counts()
```

# Encoding

# OneHotEncoding

```python
import seaborn as sns

df=sns.load_dataset('tips')
df
```

```python
smoker=len(df[df['smoker']=='Yes'])
smoker
```

```python
from sklearn.preprocessing import OneHotEncoder
oh= OneHotEncoder()

smoker_num=oh.fit_transform(df[['smoker']]).toarray()
```

```python
import pandas as pd
encoder_df=pd.DataFrame(smoker_num,columns=oh.get_feature_names_out())
encoder_df
```

```python
smoke_num=pd.concat([df,encoder_df],axis=1)
smoke_num.drop(columns=['smoker'],inplace=True)
smoke_num
```

```python
len(smoke_num[smoke_num['smoker_No']==1])
```

# Label Encoding

```python
smoke_num['day'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
days=le.fit_transform(smoke_num[['day']])
day_num=pd.DataFrame(days,columns=['day_num'])
final=pd.concat([smoke_num,day_num],axis=1)
final.drop(columns=['day'],inplace=True)
final
```

```python
final['sex'].value_counts()
```

```python
gender_le=LabelEncoder()
genders=gender_le.fit_transform(final['sex'])

sex_df=pd.DataFrame(genders,columns=['sex_df'])
sex_df
```

```python
final=pd.concat([final.drop(columns='sex'),sex_df],axis=1)
```

```python
final['time'].value_counts()
final
```

# Target Guided Ordinal Encoding

```python
final['time'].value_counts()
```

```python
a = final.groupby('time')['total_bill'].mean()
a
```

```python
final['time_encoded']=final['time'].map(a)
final
```

# Ordinal Encoding

```python
import pandas as pd

data = {
    "Person_ID": range(1, 11),
    "Education_Level": [
        "High School", "Bachelor", "Master", "Bachelor", "PhD",
        "Master", "High School", "PhD", "Bachelor", "Master"
    ],
    "Years_Experience": [1, 3, 5, 2, 10, 6, 0, 12, 4, 7],
    "Salary": [30000, 45000, 60000, 40000, 90000, 65000, 28000, 120000, 50000, 70000]
}

df = pd.DataFrame(data)
df
```

```python
df['Education_Level'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()
abc=oe.fit_transform(df[['Education_Level']],['High School','Bachelor','Master','PhD'])


df['Education_Level_Encoded']=pd.DataFrame(abc)

df.drop(columns=['Education_Level'])
```
